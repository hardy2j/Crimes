```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rape_victims=pd.read_csv('/Users/y2z/Desktop/Downloads/20_Victims_of_rape.csv.xls')
```

```python
rape_victims
```

```python
rape_victims[rape_victims['Rape_Cases_Reported']!=rape_victims['Victims_of_Rape_Total']].head()
```

```python
rape_victims['Unreported_Cases']=rape_victims['Victims_of_Rape_Total']-rape_victims['Rape_Cases_Reported']
```

```python
rape_victims[rape_victims['Area_Name']=='Delhi']
```

```python
unreported_rape_victims=rape_victims.groupby('Area_Name').sum()
unreported_rape_victims=unreported_rape_victims.sort_values(by='Victims_of_Rape_Total',ascending=False)
unreported_rape_victims.drop(['Year'],axis=1,inplace=True)
unreported_rape_victims=unreported_rape_victims[unreported_rape_victims['Unreported_Cases']>0]
plt.figure(figsize=(15,10))
ct=unreported_rape_victims[unreported_rape_victims['Unreported_Cases']>0]['Unreported_Cases'].sort_values(ascending=False)
ct.plot.bar()
plt.xticks(rotation=45)
plt.xlabel('States')
plt.ylabel('Unreported Cases')
plt.title('States VS Unreported Cases')
plt.show()
```

```python
rape_victims_by_state=rape_victims.groupby('Area_Name').sum().reset_index()
rape_victims_by_state.drop('Year',axis=1,inplace=True)
print("Total number of rape cases reported are:",rape_victims_by_state['Rape_Cases_Reported'].sum())
rape_victims_by_state=rape_victims_by_state.sort_values(by='Rape_Cases_Reported',ascending=False)
```

```python
# rape_victims_by_state
```

```python
plt.figure(figsize=(10,10))
plt.bar(x=rape_victims_by_state.Area_Name,height=rape_victims_by_state.Rape_Cases_Reported)
plt.xticks(rotation=90)
plt.xlabel("State")
plt.ylabel("No.of Cases Reported")
plt.title("State VS No.of Cases Reported")
plt.show()
```

```python
cr=rape_victims_by_state.groupby(['Area_Name'])['Rape_Cases_Reported'].sum().reset_index()
cr=cr.sort_values(by='Rape_Cases_Reported',ascending=False)
cr
```

```python
mp_rape_victims=rape_victims[rape_victims['Area_Name']=='Madhya Pradesh']
mp_rape_victims_year=mp_rape_victims.groupby('Year').sum()
mpcr=mp_rape_victims_year['Rape_Cases_Reported'].plot()
mpcr.xaxis.set_ticks(np.arange(2001,2011,1))
plt.show()
```

```python
mp_incest_rape_victims=mp_rape_victims[mp_rape_victims['Subgroup']=='Victims of Incest Rape']
mp_incest_rape_victims=mp_incest_rape_victims.groupby('Year').sum()
mp_incest_rape_victims['Rape_Cases_Reported'].plot.bar()
plt.show()
```
